```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("imgs/treino e teste.png")
plt.imshow(img)
plt.axis('off')  # Remove os eixos
plt.show()

# X_train é a feature de treino
# y_train é o target de treino
# X_test é a feature de teste
# y_treino é o target de teste
```

```python
# remover a coluna salary e transformar em numpy array
# features 
X = np.array(df_dummy.drop(['salary'], axis=1))

# target
y = np.array(df_dummy['salary'])
```

```python
# separar pra o treino e o teste
# test_size = 0.2 = 20%
# random_state = deixar as linhas aleatórias, randomizar as linhas
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
# normalizar as features pra valores dentro de um range
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)
```

```python
# quantidade de vizinhos = n_neighbors = 19
# a métrica usada = metric='euclidean'
clf = neighbors.KNeighborsClassifier(n_neighbors=19, metric='euclidean')
```

```python
# treinar usando o modelo
clf.fit(X_train, y_train)
```

```python
# ver a acuracia usando os dados de test
# accuracy = quantidade que o modelo acertou
# ou seja, tem x% a probabilidade de acertar quando eu testar com um novo valor.
accuracy = clf.score(X_test, y_test) 
porcent_accuracy = '%.2f'%(accuracy * 100) 
print(f'O modelo acerto {porcent_accuracy}% dos casos.') 
print(f'Ou seja, tem {porcent_accuracy}% a probabilidade de acertar quando eu testar com um novo valor.')
```

```python
# função pra visualizar alguns dados
def knn_comparison(data, k):
    x = data[['ssc_p', 'hsc_p']].values
    y = data['salary'].astype(int).values
    knn = neighbors.KNeighborsClassifier(n_neighbors=k)
    knn.fit(x,y)
    plot_decision_regions(x,y,clf=knn)
    plt.xlabel('ssc_p')
    plt.ylabel('hsc_p')
    plt.title('K='+str(k))
    plt.show()
```

```python
knn_comparison(data=df_dummy, k=19)
```
